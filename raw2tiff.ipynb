```python
import numpy as np
import tifffile

raw_path = r"C:\BingShiuan rotation\0129 stim BW03\1\Image_001_001_.raw"
n_frames = 6000
height = 512
width  = 512

mm = np.memmap(raw_path, dtype=np.uint16, mode="r", shape=(n_frames, height, width))

out_tif = r"C:\BingShiuan rotation\0129 stim BW03\1\movie_converted.tif"
with tifffile.TiffWriter(out_tif, bigtiff=True) as tw:
    for i in range(n_frames):
        tw.write(mm[i])
print("Wrote:", out_tif)
```
